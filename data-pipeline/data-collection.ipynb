```python
def request_proposals(governance):
    r = requests.post(snapshot_api, json={
        'query': GET_PROPOSALS, 
        'variables': {
            'first': 10000,
            'skip': 0,
            'space_in': [governance]
        }
    })
    print('Status:', r.status_code)
    return json.loads(r.text)['data']['proposals']

def request_votes(proposal_id):
    r = requests.post(snapshot_api, json={
        'query': GET_VOTES, 
        'variables': {
            'first': 100000,
            'skip': 0,
            'proposal': proposal_id
        }
    })
    print('Status:', r.status_code)
    return json.loads(r.text)['data']['votes']
```

### Proposal Collection

```python
spaces = df.loc[df['governance_type']=='Snapshot']['governance'].values
```

```python
all_data = []
for s in spaces:
    proposals = request_proposals(s)
    total_proposal = len(proposals)
    print(s, total_proposal)
    for p in proposals:
        data = [p['id'], p['title'], p['body'], p['choices'], p['start'], p['end'], p['snapshot'], p['state'], 
                p['author'], p['space']['id'], p['space']['name'],total_proposal]
        all_data.append(data)
```

```python
df = pd.DataFrame(all_data, columns=colnames)
```

```python
df.to_csv('../data/snapshot_proposals.csv', index=False)
```

### Vote Collection

```python
all_votes = []
colnames = ['id', 'voter', 'created', 'proposal', 'choice', 'space_name', 'total_votes']

p_ids = df['id'].values
for p_id in p_ids:
    votes = request_votes(p_id)
    total_votes = len(votes)
    for v in votes:
        data = [v['id'], v['voter'], v['created'], p_id, v['choice'], v['space']['id'], total_votes]
        all_votes.append(data)
```

```python
df_votes = pd.DataFrame(all_votes, columns=colnames)
df_votes.to_csv('../data/snapshot_votes.csv', index=False)
```
